```python
u = np.array([2,4,5,6])
```

```python
v = np.array([1,0,0,2])
```

```python
def vector_vector_multiplication(u,v):
    assert u.shape[0] == v.shape[0]

    n = u.shape[0]
    
    result = 0.0

    for i in range(n):
        result += u[i] * v[i]

    return result
```

```python
vector_vector_multiplication (u,v)
```

```python
u.dot(v)
```

Matrix-Vector multiplication

So if p has 4 columns, w must have 4 elements to match.

```python
p = np.array([
    [2,4,5,6],
    [1,2,1,2],
    [3,1,2,1],
])
```

```python
p.shape
```

The number of columns in p should match the number of elements in w

```python
w = np.array([1,0.5,2,1])
```

```python
w.shape[0]
```

The result vector needs to hold one number per row of the matrix and the matrix has 3 rows, so the result needs to hold 3 numbers.

Now, p.shape gives you (rows, columns), so:

p.shape[0] = number of rows = 3

So np.zeros(p.shape[0]) is simply creating an empty vector of size 3

Which will eventually be filled with your answers [20, 6, 8.5] as the loop runs

```python
def matrix_vector_multiplication (p,w):
    assert p.shape[1] == w.shape[0]

    result = np.zeros(p.shape[0])

    for i in range(p.shape[0]):
        for j in range(p.shape[1]):
            result[i] += p[i][j] * w[j]

    return result
```

```python
matrix_vector_multiplication (p,w)
```

Note the function below is different

    result[i] = vector_vector_multiplication(p[i], w)

The v_v_m function from above is reused to replace this line of code
   
   for j in range(p.shape[1]):

p[i] extracts the entire row [i] as a vector, then passes it straight into vector_vector_multiplication along with w.

Why
Matrix × Vector = just doing Vector × Vector repeatedly, once per row
    vector_vector_multiplication — reuse the already built function that handles one row × one column
    p[i] — grab one full row from the matrix at a time
    w — the single vector (one column) stays the same every time

```python
def matrix_vector_multiplication (p,w):
    assert p.shape[1] == w.shape[0]

    num_rows=p.shape[0]

    result = np.zeros(num_rows)

    for i in range(num_rows):
        result[i] = vector_vector_multiplication(p[i], w)

    return result
```

```python
matrix_vector_multiplication (p,w)
```

```python
p.dot(w)
```

Now lets do a matrix_matrix_multiplication

```python
a = np.array([
    [1,1,2],
    [0,0.5,1],
    [0,2,1],
    [2,1,0],
])
```

```python
def matrix_matrix_multiplication (p,a):
    assert p.shape[1] == a.shape[0]

    result = np.zeros((p.shape[0],a.shape[1]))

    for i in range (a.shape[1]):
        result[:,i] = matrix_vector_multiplication(p, a[:,i])

    return result
```
