Multiplication

```python
import numpy as np
```

```python
u = np.array([2,4,5,6])
```

```python
u.shape[0]
```

```python
v = np.array([1,0,0,2])
```

```python
v.shape[0]
```

```python
def vector_vector_multiplication(u,v):
    assert u.shape[0] == v.shape[0]

    n = u.shape[0]
    
    result = 0.0

    for i in range(n):
        result = result + u[i] * v[i]

    return result
```

```python
vector_vector_multiplication (u,v)
```

```python
u.dot(v)
```

Matrix-Vector multiplication

So if p has 4 columns, w must have 4 elements to match.

```python
p = np.array([
    [2,4,5,6],
    [1,2,1,2],
    [3,1,2,1],
])
```

```python
p.shape
```

The number of columns in p should match the number of elements in w

```python
w = np.array([1,0.5,2,1])
```

```python
w.shape[0]
```

```python
def matrix_vector_multiplication (p,w):
    assert p.shape[1] == w.shape[0]

    n=p.shape[1]

    result = 0.0

    for i in range(n):
        result = result + p[i] * w[i]

    return result
```

```python
matrix_vector_multiplication (p,w)
```
